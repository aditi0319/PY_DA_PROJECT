```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #visualizing data
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')#df (DATA FRAME)is like a variable to store data 
```

```python
df.shape # it will show numbers of rows and columns res.
```

```python
df.head(7) # it will show top 5 columns you can see more by giving bracket a number.
```

```python
df.info() # it is used in Data cleaning , it will show number of rows , coulumns, data type etc
```

```python
#drop unrelated/blank columns and inplace is used to implement the changes
df.drop(['Status', 'unnamed1'], axis=1, errors='ignore', inplace=True)
```

```python
pd.isnull(df) # to check null value
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace = True) # to drop null value and inplace is used to save and implement the changes
```

```python
df['Amount'] = df['Amount'].astype('int') # changed the data type from float to int
```

```python
df['Amount'].dtypes # check changed dt
```

```python
df.columns # to check whhich coulmns are there
```

```python
df.rename(columns = {'Marital_Status':'Shadi'}) # rename the column witht he help of disctionary THIS IS NOT SAVED BECAUSE WE DIDNT USE INPLACE
```

```python
df.describe() # describe() method returns description of the data in the DataFrame (i.e. count, mean, std, etc)
```

```python
# use describe() for specific columns
df[['Age', 'Orders', 'Amount']].describe()
```

```python
#NOW FROM HERE STAETED EXPLORATORY DATA ANALYSIS
```

```python
#1 GENDER
```

```python
ax = sns.countplot(x = 'Gender' ,data = df) # the ouput is graph without value
```

```python
ax = sns.countplot(x = 'Gender' ,data = df)  # the ouput is graph with value
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index= False)['Amount'].sum().sort_values (by = 'Amount' ,ascending = False)  #to calculate total amount from male anf female
```

```python
sales_gn = df.groupby(['Gender'], as_index= False)['Amount'].sum().sort_values (by = 'Amount' ,ascending = False)  #to calculate total amount from male anf female
sns.barplot(x= 'Gender', y ='Amount', data = sales_gn) #plot of amount from male and femake
```

```python
ax = sns.countplot(x= 'Age Group' ,hue = 'Gender', data = df)
for bars in ax.containers:ax.bar_label(bars)    # this part is only for giving labels
```

```python
sales_age  = df.groupby(['Age Group'] ,as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sns.barplot(x = 'Age Group', y ='Amount', data = sales_age)
```

```python
# STATE
```

```python
# total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(17,3)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
```

```python
# total amount/sales from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(17,3)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount')
```

```python
# MARITAL STATUS
```

```python
ax = sns.countplot(data = df,x = 'Marital_Status')
sns.set(rc={'figure.figsize':(6,3)})
for bars in ax.containers:
    ax.bar_label(bars)
   
```

```python
sales_state = df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(9,3)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount',hue = 'Gender')
```

```python
sns.set(rc= {'figure.figsize':(20,5)})
ax = sns.countplot(data = df,x = 'Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(15)
sns.set(rc={'figure.figsize':(15,3)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount')
```

```python
sns.set(rc={'figure.figsize':(25,5)})
ax = sns.countplot(data= df, x= 'Product_Category')
for bars in ax.containers:
     ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(15)
sns.set(rc={'figure.figsize':(15,3)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,3)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Amount')
```

```python
# Top 10 most sold products (same thing as above)
fig1, ax1 = plt.subplots(figsize = (12,4))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending= False).plot(kind='bar')
```
